```python
import cv2

def main():
    username = "admin"
    password = "Demo12345678"   # set your camera password here
    ip = "192.168.1.111"
    port = "554"    # RTSP default port

    # CP Plus RTSP stream URL
    rtsp_url = f"rtsp://{username}:{password}@{ip}:{port}/cam/realmonitor?channel=1&subtype=0"

    cap = cv2.VideoCapture(rtsp_url)

    if not cap.isOpened():
        print("Error: Cannot connect to CP Plus camera")
        return

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        cv2.imshow("CP Plus EZ-R33G Live Feed", frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

```python
import cv2

def access_camera_http():
    url = "rtsp://192.168.1.111:554/live"  # Replace with actual stream URL
    cap = cv2.VideoCapture(url)

    if not cap.isOpened():
        print("❌ Could not open video stream")
        return

    print("✅ Connected to live camera stream")

    while True:
        ret, frame = cap.read()
        if not ret:
            print("⚠️ Failed to grab frame")
            break

        cv2.imshow("Live Camera Feed", frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

access_camera_http()
```
